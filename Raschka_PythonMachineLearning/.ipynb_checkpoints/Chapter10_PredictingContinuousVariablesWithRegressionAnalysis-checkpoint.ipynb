#### Introducing a simple linear regression model

Simple linear regression can be understood as the process of finding the weights $w_{0}$ and $w_{1}$ when there is a single explanatory variable:

$$ y = w_{0} + w_{1}x $$

This process can be understood as finding the line of best fit, also known as the **regression line** for our sample set. The vertical offsets from that line to individual data points are known as **offsets** or **residuals** and represent the *error* of our linear model.

For linear regression to be useful, we must generalize to a formula that encompasses a data set with $m$ explanatory variables. Our simplified model above generalizes to **multiple linear regression** as:

$$y = w_{0}x_{0} + w_{1}x_{1} + ... + w_{m}x_{m} = \sum_{i=0}^{m}w_{i}x_{i} = \textbf{w}^{T}\textbf{x}$$

where $\textbf{w}^{T}$ is the transpose of a vector of derived weights and $\textbf{x}$ is the vector of predictors (input variables).

To judge the linear relationships among our variables, we will construct a **correlation matrix**. This matrix contains the **Pearson product-moment correlation coefficients**, which represent the covariance between two features $x$ and $y$ divided by the product of their standard deviations.

$$r = \frac{\sum_{i=1}^{n}[(x^{(i)} - \mu_{x})(y^{(i)}-\mu_{y})]}{\sqrt{\sum_{i=1}^{n}(x^{(i)} -\mu_{x})^{2}}\sqrt{\sum_{i=1}^{n}(y^{(i)} -\mu_{y})^{2}}} = \frac{\sigma_{xy}}{\sigma_{x}\sigma_{y}}$$

where $\mu$ is the mean of the sample feature, $\sigma_{xy}$ is the covariance between features $x$ and $y$, and $\sigma$ is standard deviations.

We can use numpy's *corrcoef* function and seaborn to make a heatmapped correlation matrix.

```python
cols = ['LSTAT', 'INDUS', 'NOX', 'RM', 'MEDV']
cm = np.corrcoef(df[cols].values.T)
sns.set(font_scale=1.5)
custom_style = {'axes.labelcolor': 'white',
                'xtick.color': 'white',
                'ytick.color': 'white',
                'figure.facecolor': 'black'}

sns.set_style("darkgrid", rc=custom_style)
sns.set_context(rc={"font.size":14,"axes.titlesize":14,"axes.labelsize":20})
hm = sns.heatmap(cm,
                cbar=True,
                annot=True,
                square=True,
                fmt='.2f',
                annot_kws={'size':15},
                yticklabels=cols,
                xticklabels=cols)
plt.show()
```

#### Implementing an ordinary least squares linear regression model

**Ordinary Least Squares (OLS)***

The cost function we minimized in exploring the Adaptive Linear Neuron (Adaline) in Chapter 2 was the **Sum of Squared Errors (SSE)**, which is identical to the **Ordinary Least Squares** cost function:

$$J(w) = \frac{1}{2}\sum_{i=1}^{n}(y^{(i)} - \hat{y}^{(i)})^{2}$$

where $\hat(y)$ is the predicted values $\hat(y) = w^{T}x$

The OLS linear regression can be understood as the Adaline without the unit step function. This leads to finding continuous target variables rather than binary classification outsomes (-1 or 1). Let's code the GD implementation of Adaline to illustrate this similarity.

```python
class LinearRegressionGD(object):
    
    def __init__ (self, eta=0.001, n_iter=20):
        self.eta = eta
        self.n_iter = n_iter
        
    def fit (self, X, y):
        self.w_ = np.zeros(1 + X.shape[1])
        self.cost_ = []
        
        for i in range(self.n_iter):
            output = self.net_input(X)
            errors = (y - output)
            self.w_[1:] += self.eta * (X.T.dot(errors))
            self.w_[0] += self.eta * errors.sum()
            cost = (errors**2).sum()
            self.cost_.append(cost)
        return self
    
    def net_input (self, X):
        return np.dot (X, self.w_[1:]) + self.w_[0]
    
    def predict (self, X):
        return self.net_input (X)
```

Let's test our linear regressor class by trying to predict *MEDV* using the field *RM* (number of rooms)

```python
X = df[['RM']].values
y = df['MEDV'].values
```

```python
from sklearn.preprocessing import StandardScaler

sc_x = StandardScaler()
sc_y = StandardScaler()

X_std = sc_x.fit_transform(X)
y_std = sc_y.fit_transform(y)

lr = LinearRegressionGD()
lr.fit(X_std.reshape(-1,1), y_std)
```

Below, Sebastian assesses the performance of the regressor by plotting the cast as a function of the number of iterations it's gone through. This is a good idea when trying to optimize an algorithm. We see that the regressor converged after 5 epochs.

```python
plt.plot (range(1, lr.n_iter+1), lr.cost_)
plt.ylabel('SSE')
plt.xlabel('Epoch')
plt.show()
```

Now, we can see how well the regressor did finding the line of best fit.

```python
def lin_regplot (X, y, model):
    plt.scatter(X, y, c='purple')
    plt.plot(X, model.predict(X), color='orange')
    return None
```

```python
lin_regplot(X_std, y_std, lr)
plt.xlabel('Average number of rooms [RM] (standardized)')
plt.ylabel('Price in $1000\'s [MEDV] (USD; standardized)')
plt.show()
```

Though this data can be generally interpreted to say that housing prices increase with the number of rooms, it clearly doesn't explain things very well. We also see a curious line of points at y=3, meaning prices may have been clipped.

We can project our predicted price outcome back onto our original scale to get a more meaningful predicted value.

```python
num_rooms_std = sc_x.transform([5.0])
price_std = lr.predict(num_rooms_std)
print("Price in 1000's: %.3f" % \
         sc_y.inverse_transform(price_std))
```

Interesting sidenote from Raschka: one can solve the OLS in a closed-form using a system of linear equations:

$$w = (X^{T}X)^{-1}X^{T}y$$

In code form:

```python
# adding a column vector of "ones"

Xb = np.hstack((np.ones((X.shape[0], 1)), X))
w = np.zeros(X.shape[1])
z = np.linalg.inv(np.dot(Xb.T, Xb))
w = np.dot(z,np.dot(Xb.T, y))
print('Slope: %.3f' % w[1])
print('Intercept: %.3f' % w[0])
```

This method will find the optimal solution analytically, but will be computationally expensive for larger datasets.

#### Fitting a robust regression model using RANSAC

Regression models are very sensitive to outliers. Detecting these small subsets of data that can negatively impact our models is beyond the scope of Raschka's book, but he discusses how to implement something known as **RANdom SAmple Consensus (RANSAC)**, a robust implementation of linear regression that fits a model to *inliers*.

What follows are the basic steps to implementing RANSAC:

1. Select a random number of samples to be inliers and fit the model
2. Test all other data points against the fitted model and add those points that fall within a user-given tolerance to the inliers
3. Refit the model using all inliers
4. Estimate the error of the fitted model vs inliers.
5. Terminate the algorithm if performance meets a certain user-defined threshold or if a fixed number of iterations has been reached; go back to step 1 otherwise.

```python
from sklearn.linear_model import RANSACRegressor

ransac = RANSACRegressor(LinearRegression(),
                        max_trials=100,
                        min_samples=50,
                        loss='absolute_loss',
                        residual_threshold=2.5,
                        random_state=0)

ransac.fit(X, y)
```

By default, scikit-learn will use **Median Absolute Deviation (MAD)** to select the inlier threshold by considering the MAD of the target vaues y.

*However, the choice of inlier threshold is very problem-specific, which is a big disadvantage of RANSAC*

For more discussion on determining inlier thresholds, check out [this paper](https://github.com/devellocus/kingsmen-analytics/blob/master/MachineLearningResources/Literature/AutomaticEstimationoftheInlierThreshold_Toldo_Fusiello.pdf) by R. Toldo and A. Fusiello.

```python
inlier_mask = ransac.inlier_mask_
outlier_mask = np.logical_not(inlier_mask)
line_X = np.arange(3, 10, 1)
line_y_ransac = ransac.predict(line_X[:, np.newaxis])
sns.set_context(font_scale=1.5, rc={"font.size":20,"axes.titlesize":14,"axes.labelsize":20, 
                    'legend.frameon': True, 'legend.labelsize': 20})
plt.scatter(X[inlier_mask], y[inlier_mask],
           c='green', marker= 'o', label='Inliers')
plt.scatter(X[outlier_mask], y[outlier_mask],
           c='red', marker='x', label='Outliers')
plt.plot(line_X, line_y_ransac, color='blue')
plt.xlabel('Average number of rooms [RM]')
plt.ylabel('Price in 1000\'s [MEDV]')
plt.legend(loc='upper left')
plt.show()
```

```python
print('Slope: %.3f' % ransac.estimator_.coef_[0])
print('Intercept: %.3f' % ransac.estimator_.intercept_)
```

Note that the book used a different value for inlier threshold. I tried to obtain results that agreed with the book using 5.0 as the threshold, but I obtain different results. This is likely due to the updated version of RANSACRegressor, which changes the method of specifying the residual_metric, using built-in 'loss' metrics instead of user-defined lambdas.

#### Evaluating the performance of linear regression models

Thus far, we have not cared about splitting on training and test data, but when building models it's incredibly important. As you may remember from previous sections, overfitting will plague a model built and tested on its training data.

Also, let's try buildiong our model on all of the features, rather than just RM. Now, this problem becomes one of **multiple regression**.

```python
from sklearn.cross_validation import train_test_split

X = df.iloc[:, :-1].values
y = df['MEDV'].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
slr = LinearRegression()
slr.fit(X_train, y_train)
y_train_pred = slr.predict(X_train)
y_test_pred = slr.predict(X_test)
```

Our model is trained on a dataset of dimensionality > 2, so scattering the variables against our target variable is not possible with a 2D plot. Instead we will plot the **residuals**, which are the differences between actual and predicted values, against the predicted values themselves. **Residual plots** like these are common for diagnosing issues in a regression model.

```python
plt.style.use('dark_background')
plt.scatter(y_train_pred, y_train_pred - y_train,
            c='purple', marker='o', label='Training data')
plt.scatter(y_test_pred, y_test_pred - y_test,
            c='orange', marker='s', label='Test data')
plt.xlabel('Predicted values')
plt.ylabel('Residuals')
plt.title('Housing Data Linear Regression - Residuals')
plt.hlines(y=0, xmin=-10, xmax=50, lw=5, color='white',)
plt.xlim([-10,50])
plt.show()
```

###### Interpreting a residual plot

A perfect model would have all residuals = 0, which will not happen in reality. A good model will have residuals randomly scattered about the y=0 centerline. Patterns in the scattering indicate a failure on the part of the model to capture some explanatory information in the feature set.

We see a linear relationship in the bottom right. This would indicate our model has failed to capture some explanatory variable. Additionally, outlier points can be detected as points with high displacements from the centerline

We can also calculate the **Mean Squared Error** of our model, which is the average value of the *Sum of Squared Errors (SSE)*

$$MSE = \frac{1}{n}\sum_{i=1}^{n}(y^{(i)}-\hat{y}^{(i)})^{2}$$

```python
from sklearn.metrics import mean_squared_error
print('MSE train: %.3f' % mean_squared_error(y_train, y_train_pred))
print('MSE test: %.3f' % mean_squared_error(y_test, y_test_pred))
```

A significantly higher MSE on the test set indicates overfitting. We can also look at another useful metric, the **coefficient of determination ($R^{2}$)**.

$$R^{2} = 1 - \frac{MSE}{Var(y)}$$

This value is bounded between 0 and 1 for the training data, but can be negative for the test set.

```python
from sklearn.metrics import r2_score

print('R^2 train: %.3f' % r2_score(y_train, y_train_pred))
print('R^2 test: %.3f' % r2_score(y_test, y_test_pred))
```

#### Using regularized methods for regression

We can tackle problems with overfitting using methods of *regularizing* our dataset. as we explored in Ch. 3. Some popular regularization methods for linear regression tasks are:

- **Ridge Regression**
- **Least Absolute Shrinkage and Selection Operator (LASSO)**
- **Elastic Net**

###### Ridge Regression

This is a L2 regularized model with an additive regularization term to the lease-squares cost function.

$$ J(w)_{Ridge} = \sum_{i=1}^{n}(y^{(i)} - \hat{y}^{(i)})^2 + \lambda||w||_{2}^{2}$$

Where:

$$L2 : \lambda||w||_{2}^{2} = \lambda\sum_{j=1}^{m}w_{j}^{2}$$

The L2 term is the *regularization strength*, which shrinks the weights of our model as we increase it.

###### Least Absolute Shrinkage and Selection Operator (LASSO)

LASSO is very useful for supervised feature selection because it has a tendency to force certain weights to zero. This is due to the fact that it uses an L1 regularization parameter. It will often lead to sparse models.

$$J(w)_{LASSO} = \sum_{i=1}^{n}(y^{(i)} - \hat{y}^{(i)})^{2} + \lambda||w||_{1}$$

Where:

$$L1: \lambda||w||_{1} = \lambda\sum_{j=1}^{m}|w_{j}|$$

###### Elastic Net

A nice compromise between Ridge and LASSO is the Elastic Net, which has both L1 and L2 terms to generate sparsity while overcoming LASSO's inability to select more than n < m features.

$$ J(w)_{ElasticNet} = \sum_{i=1}^{n}(y^{(i)} - \hat{y}^{(i)})^2 + \lambda||w||_{2}^{2} + \lambda||w||_{1}$$

These are all implemented in scikit-learn, but we have to specify the regulatization strength $\lambda$, called *alpha* in the sklearn implementations.

```python
from sklearn.linear_model import Ridge, Lasso, ElasticNet

ridge = Ridge(alpha=1.0)
lasso = Lasso(alpha=1.0)
elastic_net = ElasticNet(alpha=1.0, l1_ratio=0.5)
```

We see that we vary the L1 to L2 ratio when using the elastic net. It should also be noted that alpha is not quite the same as $\lambda$ in practice, so understanding the difference is important. Documentation can be found [here](http://scikit-learn.org/stable/modules/linear_model.html).

#### Turning a linear regression problem into a curve - polynomical regression

We can account for violation of linearity in a dataset by adding polynomial terms to create a **polynomial regression model**

$$y = w_{0} + w_{1}x + w_{2}x^{2} + ... + w_{d}x^{d}$$

The problem is still considered a multiple *linear* regression problem because the coefficients $w_{i}$ are linear. We can add the polynomial term to our linear regression model using the **PolynomialFeatures** transformer class in scikit-learn.

###### 1. Add a second degree polynomial term:

```python
from sklearn.preprocessing import PolynomialFeatures

X = np.array([258.0, 270.0, 294.0, 320.0, 342.0, 368.0, 396.0, 446.0, 480.0, 586.0])[:, np.newaxis]
y = np.array([236.4, 234.4, 252.8, 298.6, 314.2, 342.2, 360.8, 368.0, 391.2, 390.8])


quadratic = PolynomialFeatures(degree=2)
X_quad = quadratic.fit_transform(X)
```

###### 2. Fit a simple linear regression model for comparison:

```python
lr = LinearRegression()
lr.fit(X, y)
X_fit = np.arange(250,600,10)[:, np.newaxis]
y_lin_fit = lr.predict(X_fit)
```

###### 3. Fit a multiple regression model on the transform features for polynomial regression:

```python
pr = LinearRegression()
pr.fit(X_quad, y)
y_quad_fit = pr.predict(quadratic.fit_transform(X_fit))
```

###### 6. Plot the results:

```python
plt.style.use('dark_background')
plt.scatter(X, y, label='training points', color='white')
plt.plot(X_fit, y_lin_fit,
            color='lightblue',label='linear fit', linestyle='--')
plt.plot(X_fit, y_quad_fit,
            color='red', label='quadratic fit')
plt.legend(loc='upper left')
plt.show()
```

We can see from the plot that the quadratic fit captures the relationship much better than the linear fit, but let's prove it by assessing the model's respective performances.

```python
y_lin_pred = lr.predict(X)
y_quad_pred = pr.predict(X_quad)

print('Training MSE linear: %.3f' % mean_squared_error(y, y_lin_pred))
print('Training MSE quadratic: %.3f' % mean_squared_error(y, y_quad_pred))
print('Training R^2 linear: %.3f' % r2_score(y, y_lin_pred))
print('Training R^2 linear: %.3f' % r2_score(y, y_quad_pred))
```

#### Modeling nonlinear relationships in the Housing Dataset

Raschka outputs a nice bit of code here to fit models on multiple polynomial levels using LSTAT to predict home prices from the original dataset.

```python
X = df[['LSTAT']].values
y = df['MEDV'].values
regr = LinearRegression()

#create polynomial features
quadratic = PolynomialFeatures(degree=2)
cubic = PolynomialFeatures(degree=3)
X_quad = quadratic.fit_transform(X)
X_cubic = cubic.fit_transform(X)

#linear fit
X_fit = np.arange(X.min(), X.max(), 1)[:, np.newaxis]
regr = regr.fit(X, y)
y_lin_fit = regr.predict(X_fit)
linear_r2 = r2_score(y, regr.predict(X))

#quadratic fit
regr = regr.fit(X_quad, y)
y_quad_fit = regr.predict(quadratic.fit_transform(X_fit))
quadratic_r2 = r2_score(y, regr.predict(X_quad))

#cubic fit
regr = regr.fit(X_cubic, y)
y_cubic_fit = regr.predict(cubic.fit_transform(X_fit))
cubic_r2 = r2_score(y, regr.predict(X_cubic))

#plot results
plt.scatter(X, y, label='training points', color='white', alpha=0.4)
plt.plot(X_fit, y_lin_fit, label = 'linear (d=1), $R^2=%.2f$' % linear_r2,
            color='lightgreen', lw=4, linestyle=':')
plt.plot(X_fit, y_quad_fit, label = 'quadratic (d=2), $R^2=%.2f$' % quadratic_r2,
            color='purple', lw=3, linestyle='-')
plt.plot(X_fit, y_cubic_fit, label = 'cubic (d=3), $R^2=%.2f$' %cubic_r2,
            color='orange', lw=3, linestyle='--')
plt.xlabel('% lower status of the population [LSTAT]')
plt.ylabel('Price in $1000\'s [MEDV]')
plt.title('Nonlinear relationships in the housing dataset', fontsize=30)
plt.legend(loc='upper right', fontsize=20)
plt.show()
```

We see that the cubic fit was best. It should be noted however, that adding more degrees of polynomial features adds to model complexity and increases the likelihood of overfitting.

To demonstrate that polynomial features are not a catch-all solution, Raschka does another transform by surmising that a logarithmic transformation of LSTAT and taking the square root of MEDV might project the dataset onto a feature space that works better with linear regression

```python
#transform features
X_log = np.log(X)
y_sqrt=np.sqrt(y)

#fit features
X_fit = np.arange(X_log.min()-1,
                  X_log.max()+1, 1)[:,np.newaxis]
regr = regr.fit(X_log, y_sqrt)
y_lin_fit = regr.predict(X_fit)
linear_r2 = r2_score(y_sqrt, regr.predict(X_log))

#plot features
plt.scatter(X_log, y_sqrt, label='training points', color='white', alpha=0.6)
plt.plot(X_fit, y_lin_fit, label='linear (d=1), $R^2=%.2f$' % linear_r2, color='lightblue', lw=3)
plt.xlabel('log(% lower status of the population) [LSTAT]')
plt.ylabel('$\sqrt{Price \; in \; \$1000\'s [MEDV]}$')
plt.legend(loc='lower left', fontsize=20)
plt.title('Mapping housing dataset to other non-polynomial dimensional spaces', fontsize=30)
plt.show()
```

This seems to have done an even better job at capturing the relationship.

### Dealing with nonlinear relationships using random forest

We move now to a fundamentally different type of regression: **random forest regression**. This model uses an ensemble of **decision trees** to create a model that is based on a *sum of piecewise linear functions*. This subdivides the input space into smaller, more manageable regions.

#### Decision tree regression

If you'll recall, when we learned about pruning decision trees in chapter 3, we sought to maximize *Information Gain* by iteratively splitting our data sets until the leaves were "pure", where *entropy* was our measure of a leaf's impurity. We defined information gain for binary classification tasks as:

$$IG(D_{p}, x_{i}) = I(D_{p}) - \frac{N_{left}}{N_{p}}I(D_{left}) - \frac{N_{right}}{N_{p}}I(D_{right})$$

We seek to do the same thing for regression tasks - maximize information gain and reduce impurity in child nodes. However, we move from using entropy as our measure of impurity of a node t to *MSE*:

$$I(t) \; = \; MSE(t) = \frac{1}{N_{t}}\sum_{i\in{D_{t}}}(y^{(i)} - \hat{y}_{t})^{2}$$

- $N_{t}$ is the number of training samples at node $t$
- $D_{t}$ is the training subset at node $t$
- $y^{(i)}$ is the true target value
- $\hat{y}_{t}$ is the predicted target value $$\hat{y}_{t}=\frac{1}{N}\sum_{i\in{D_{t}}}y^{(i)}$$

MSE in this context is often referred to as within-node variance, so we call our splitting criterion *variance reduction*. Let's use scikit-learn's DecisionTreeRegressor class to model the nonlinear relationship between **MEDV** and **LSTAT**.

```python
from sklearn.tree import DecisionTreeRegressor

X = df[['LSTAT']].values
y = df['MEDV'].values
tree = DecisionTreeRegressor(max_depth=3)
tree.fit(X, y)
sort_idx = X.flatten().argsort()
lin_regplot(X[sort_idx], y[sort_idx], tree)
plt.xlabel('% lower status of the population [LSTAT]')
plt.ylabel('Price in 1000\'s [MEDV]')
plt.show()
```

Though the model fits the trend fairly well, it does not preserve differentiability of the data. One must also be careful about choosing max_depth: choosing a poor value can lead to over/underfitting.

#### Random Forest Regression

Like we discussed in Chapter 3, a random forest uses an ensemble of decision trees to fit a robust model. Using a random forest over a single decision tree leads to a model that generalizes to other datasets better and is less sensitive to outliers.

Let's train a Random Forest regressor on all features in the Housing Dataset. The training set will be 60% of the entire dataset, and we'll evaluate on the other 40%.

```python
X = df.iloc[:, :-1].values
y = df['MEDV'].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.4, random_state=1)

from sklearn.ensemble import RandomForestRegressor

forest = RandomForestRegressor(n_estimators=1000, criterion='mse', random_state=1, n_jobs=-1)
forest.fit(X_train, y_train)
y_train_pred = forest.predict(X_train)
y_test_pred = forest.predict(X_test)

print('MSE train: %.3f' % mean_squared_error(y_train, y_train_pred))
print('MSE test: %.3f' % mean_squared_error(y_test, y_test_pred))
print('$R^{2}$ train: %.3f' % r2_score(y_train, y_train_pred))
print('$R^{2}$ test: %.3f' % r2_score(y_test, y_test_pred))
```

Unfortunately, the random forest is overfitting the data, though it is still able to explain the relationship fairly well. Let's plot the residuals.

```python
plt.scatter(y_train_pred,
            y_train_pred - y_train,
            c='white',
            marker='o',
            s=35,
            alpha=0.4,
            label='Training data')
plt.scatter(y_test_pred,
            y_test_pred - y_test,
            c='lightblue',
            marker='s',
            s=35,
            alpha=0.8,
            label='Test data')
plt.xlabel('Predicted values')
plt.ylabel('Residuals')
plt.legend(loc='upper left')
plt.hlines(y=0, xmin=-10, xmax=50, lw=4, color='lightgreen')
plt.xlim([-10,50])
plt.show()
```

There's a similar story here to the earlier residual plot of the linear model. The residuals are not randomly distributed around the centerline, indicating that the model has failed to capture all of the explanatory information.
