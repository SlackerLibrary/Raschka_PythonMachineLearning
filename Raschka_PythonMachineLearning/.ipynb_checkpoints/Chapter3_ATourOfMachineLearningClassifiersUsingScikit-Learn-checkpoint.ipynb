We'll be delving into the popular scikit-learn API in this chapter to show how simple it is to implement the perceptron class we build from scratch in the previous chapter, as well as a variety of other useful classifiers.

We'll be using the Iris dataset again for its simplicity. As it so happens, the Iris dataset is contained within the scikit-learn API.

```python
from sklearn import datasets
import numpy as np
iris = datasets.load_iris()
X = iris.data[:, [2,3]]
y = iris.target
```

Note that we're only using two featurs from the dataset once more.

The Iris flower class names (e.g. *Iris-Setosa*) are stored as integers in the dataset for optimization purposes.

Unlike in our last implementation of the perceptron class, this time we will split our data into a train and test set, using built-in functionality with scikit-learn.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
            X, y, test_size=0.3, random_state=0)
```

This will split the data 70-30.

Another process that is much easier in scikit-learn is feature scaling. Here, we'll use the **StandardScaler** class to scale our data.

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
sc.fit(X_train)
X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)
```

**StandardScaler** estimated the mean and st dev of the data sets and transformed them using those parameters.

We'll train a perceptron now. scikit-learn algorithms have default support for multiclass classification via **One-vs-Rest** methodology (see Raschka pg 28).

```python
from sklearn.linear_model import Perceptron
ppn = Perceptron(n_iter=40, eta0=0.01, random_state=0)
ppn.fit(X_train_std, y_train)
```

Note that the notation here is very similar to the Perceptron we coded in the previous chapter. We can also use a **predict** method within sklearn's **Perceptron** class just like we would have with our hand-coded example.

```python
y_pred = ppn.predict(X_test_std)
print('Misclassified samples: {}'.format((y_test != y_pred).sum()))
```

```python
misclassification_rate = ((y_test != y_pred).sum()) / len(y_test)
accuracy = 1 - misclassification_rate
print(accuracy)
```

As we see above, we can define a term *accuracy* that is 1 - the rate at which our model misclassfies test samples.

sklearn also has many native performance metrics we can take a look at.

```python
from sklearn.metrics import accuracy_score
print('Accuracy: {}'.format(accuracy_score(y_test, y_pred)))
```

We'll be implementing the *plot_decision_regions* method we created in the last chapter. However, we'll modify it to draw circles around the **test** data in order to differentiate it.

```python
from matplotlib.colors import ListedColormap
import matplotlib.pyplot as plt
```

```python


def plot_decision_regions(X, y, classifier, 
                          test_idx=None,resolution=0.02):
    #setup marker generator and color map
    markers = ('s', 'x', 'o', '^', 'v')
    colors = ('red', 'blue', 'lightgreen','gray', 'cyan')
    cmap = ListedColormap(colors[:len(np.unique(y))])
    
    #plot the decision surface
    x1_min, x1_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    x2_min, x2_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution),
                           np.arange(x2_min, x2_max, resolution))
    
    Z = classifier.predict(np.array([xx1.ravel(), xx2.ravel()]).T)
    Z = Z.reshape(xx1.shape)
    
    plt.contourf(xx1, xx2, Z, alpha=0.4, cmap=cmap)
    plt.xlim(xx1.min(), xx1.max())
    plt.ylim(xx2.min(), xx2.max())
    
    #plot class samples
    f, (ax1, ax2)
    for idx, c1 in zip(enumerate(np.unique(y)),:
        plt.scatter(x=X [y == c1, 0], y=X [y == c1, 1],
                   alpha=0.8, c=cmap(idx),
                   marker=markers [idx], label=c1)
        
    #highlight test samples
    if test_idx:
        X_test, y_test = X[test_idx,:], y[test_idx]
        plt.scatter(X_test[:, 0], X_test[:, 1], c='', edgecolors='w',
                    alpha=1.0, linewidths=1, marker='o',
                    s=55, label='test set')
```

Now we just need to combine our test and train data back together and tell our *plot_decision_regions* method what the range of indices contains our test data.

```python
X_combined_std = np.vstack((X_train_std, X_test_std))
y_combined = np.hstack((y_train,y_test))

plt.style.use('dark_background')

plot_decision_regions(X=X_combined_std,
                      y=y_combined,
                      classifier=ppn,
                      test_idx=range(105,150))

plt.xlabel('petal length [standardized]')
plt.ylabel('petal width [standardized]')
plt.legend(loc='upper left')
plt.show()
```

Here is a situation where our data is **not** linearly separable. This is why the use of the perceptron is not recommended in practice as it will fail to converge.

Thus, we have a need to find a linear classifier that *will* converge when data is not linearly separable.

Despite its name, *logistic regression* is a classification algorithm that performs very well on linearly seperable data. Also, its cost function can converge even if the data is not linearly seperable.

To predict the probability that a certain sample belongs to a particular class, we apply the sigmoid function to our net input **z**:

*The sigmoid function*
$$\phi(z)=\frac{1}{1+e^{-z}}$$ 

We'll plot the sigmoid function for values from -7 to 7 just to get a feel for what it looks like.

```python
import matplotlib.pyplot as plt
import numpy as np

def sigmoid(z):
    return 1.0 / (1.0 + np.exp(-z))

z = np.arange(-7, 7, 0.1)
phi_z = sigmoid(z)

plt.xlabel('z')
plt.ylabel('$\phi (z)$')

plt.style.use('dark_background')

plt.plot(z, phi_z)
plt.axvline(0.0, color='w')
plt.axhspan(0.0, 1.0, facecolor='0.0', edgecolor='w',alpha=1.0, ls='dotted')
plt.axhline(y=0.5, ls='dotted', color='w')
plt.yticks([0.0, 0.5, 1.0])
plt.ylim(-0.1, 1.1)
plt.show()
```

Some takeaways: 
* $\phi(z)$ approaches 1 if z goes towards $\infty$
    * $e^{-z}$ becomes very small for large values of *z*
* $\phi(z)$ goes to 0 as *z* goes towards $-\infty$
    * The denominator of the sigmoid function goes to $\infty$

*The sigmoid takes real value inputs and transforms them to values in the range of [0, 1] with an intercept at $\phi(z) = 0.5$*

If we use the sigmoid function in our *quantizer*, using the knowledge that our intercept is at $\phi(z) = 0.5$, we write our quantizer as follow: $$\hat{y} = 1 \textit{   if z $\geq$ 0}$$ and is zero otherwise

From this we define the *cost function J(**w**)* for logistic regression as:
$$J(\textit{w}) = \sum\limits_{i=1}^n[-y^{(i)}log(\phi(z^{(i)})) - (1 - y^{(i)})log(1-\phi(z^{(i)}))]$$

Technically, we could just replace the cost function in Adaline with our cost function we've defined here. But scikit-learn also has a logistic regression model built in, so we'll look at that.

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(C=1000.0, random_state=0)
lr.fit(X_train_std, y_train)

plt.style.use('dark_background')

plot_decision_regions(X_combined_std,
                      y_combined,
                      classifier=lr,
                      test_idx=range(105,150))

plt.xlabel('petal length [standardized]')
plt.ylabel('petal width [standardized]')
plt.legend(loc='upper left')
plt.show()

```

Let's briefly look at a prediction metric within the Logistic regression class:

```python
lr.predict_proba(X_test_std[0,:].reshape(1,-1))
```

An exceedingly important concept in machine learning is that of *regualrization*. Without careful procedure, *overfitting* can prevent models from being extensible to other data sets, and *underfitting* can prevent a model from being useful at all.

*L2 regularization* adds a term to the error function of the form: 
$$\frac{\lambda}{2}||\textbf{w}||^{2} = \frac{\lambda}{2}\sum\limits_{j=1}^mw_{j}^{2}$$
This is related to the **C** term we passed into our logistic regression function by the following:
$$C = \frac{1}{\lambda}$$

Let's gain an intution for what this regularization term is actually doing.

```python
weights, params = [], []

for c in np.arange(-5, 5):
    lr = LogisticRegression(C=10**c, random_state=0)
    lr.fit(X_train_std, y_train)
    weights.append(lr.coef_[1])
    params.append(10**c)
    
weights = np.array(weights)

plt.style.use('dark_background')

plt.plot(params, weights[:, 0],
            label = 'petal length')
plt.plot(params, weights[:, 1],
            label = 'petal width', linestyle='--')

plt.ylabel('weight coefficient')
plt.xlabel('C')
plt.legend(loc='upper left')
plt.xscale('log')
plt.show()
```

As *C* gets small (so $\lambda$ is getting large, regularization is stronger), our weight coefficients tend towards zero.

Another type of classifier is the **Linear Support Vector Machine**. Here, we are tuning the width of positive and negative hyperplanes that run parallel to the decision boundary in order to properly classify the maximum amount of data.

Similarly to our **logistic regression** model, we can pass a *C* term into our **SVC** class that serves a regularization purpose. Here, we tune the "width" of the "band" formed by the hyperplanes encompassing the decision boundary. Let's see it in action:

```python
from sklearn.svm import SVC

svm = SVC(kernel='linear', C=1.0, random_state=0)
svm.fit(X_train_std, y_train)

plt.style.use('dark_background')

plot_decision_regions(X_combined_std,
                      y_combined,
                      classifier=svm,
                      test_idx=range(105,150))

plt.xlabel('petal length [standardized]')
plt.ylabel('petal width [standardized]')
plt.legend(loc='upper left')
plt.show()
```

Our plot looks very similar to what we obtained from Logistic Regression. In practice, these two algorithms yield similar results. Some key differences:
* Logistic regression is more sensitive to outliers
* Logistic regression is simpler to update and execute, making it better for streaming data
* SVMs are easily kernelized to handle nonlinear classification problems.

To see a kernel SVM in action, we'll create a dataset representing an XOR gate using NumPy.

```python
np.random.seed(0)

X_xor = np.random.randn(200,2)
y_xor = np.logical_xor(X_xor[:, 0] > 0, X_xor[:, 1] > 0)
y_xor = np.where(y_xor, 1, -1)

plt.style.use('dark_background')

plt.scatter(X_xor[y_xor==1, 0], X_xor[y_xor==1, 1],
            c='b', marker='x', label='1')
plt.scatter(X_xor[y_xor==-1, 0], X_xor[y_xor==-1, 1],
            c='r', marker='s', label='-1')
plt.ylim(-3.0)
plt.legend()
plt.show()
```

In order to build a linear SVM that contend with this data, we have to transform the dataset to a higher dimensional space. To save us the computational expense of solving the quadratic programming task of the SVM, we replace the costly dot product as follows:
$$k(\textbf{x}^{(i)}, \textbf{x}^{(j)}) = \phi(\textbf{x}^{(i)})^{T}\phi(\textbf{x}^{(j)})$$

This is known as the *kernel trick*, where *k* is a kernel function that allows us to do this operation by transforming our dataset to a higher dimensional space.

One of the more common kernel functions is the Gaussian kernel:

$k(\textbf{x}^{(i)}, \textbf{x}^{(j)}) = exp(-\gamma||\textbf{x}^{(i)}-\textbf{x}^{(j)}||^{2}) \textit{   where   } \gamma = \frac{1}{2\sigma^{2}} \textit{   is a free parameter to be optimized}$

Due to the exponentiality of the kernel, very similar samples will result in the kernel heading towards one, whereas highly dissimilar samples result in the kernel heading towards 0.

```python
svm = SVC(kernel='rbf', random_state=0, gamma=0.10, C=10.0) #rbf stands for Radial Basis Function (Gaussian)
svm.fit(X_xor, y_xor)

plt.style.use("dark_background")

plot_decision_regions(X_xor, y_xor, classifier=svm)
plt.legend(loc='upper left')
plt.title('$\gamma$ = 0.1')
plt.show()
```

```python
svm = SVC(kernel='rbf', random_state=0, gamma=10, C=10.0) #rbf stands for Radial Basis Function (Gaussian)
svm.fit(X_xor, y_xor)

plt.style.use("dark_background")

plot_decision_regions(X_xor, y_xor, classifier=svm)
plt.legend(loc='upper left')
plt.title('$\gamma$ = 10')
plt.show()
```

```python
svm = SVC(kernel='rbf', random_state=0, gamma=100, C=10.0) #rbf stands for Radial Basis Function (Gaussian)
svm.fit(X_xor, y_xor)

plt.style.use("dark_background")

plot_decision_regions(X_xor, y_xor, classifier=svm)
plt.legend(loc='upper left')
plt.title('$\gamma$ = 100')
plt.show()
```

```python
svm = SVC(kernel='rbf', random_state=0, gamma=.01, C=10.0) #rbf stands for Radial Basis Function (Gaussian)
svm.fit(X_xor, y_xor)

plt.style.use("dark_background")

plot_decision_regions(X_xor, y_xor, classifier=svm)
plt.legend(loc='upper left')
plt.show()
```

The graphs above should give an intutive understanding of gamma. Too high, and you overfit. Too low, and it will misclassify a lot of data. $\gamma = 0.1$ was a good fit for this model.

Moving on to another type of classifier, the **decision tree** classifies data by splitting it iteratively based on how a sample responds to a series of questions (think 20 questions, the game). We split on the node with the *highest information gain*.

There are three ways of ensuring maximal information gain: **entropy, Gini impurity**, and **classification error**. The latter two are preferred for training a decision tree, but let's look at a plot of the three methods regardless. Also, we've scaled entropy down for clarity.

*Note: the mathematical rigor of examining the differences in these impurity types is expounded upon in Raschka, but I will not do it here.*

```python
def gini(p):
    return (p)*(1 - (p)) + (1-p) * (1 - (1-p))
def entropy(p):
    return -p*np.log2(p) - (1-p)*np.log2((1-p))
def error(p):
    return 1 - np.max([p, 1-p])

x = np.arange(0.0, 1.0, .01)
ent = [entropy(p) if p != 0 else None for p in x]
sc_ent = [e * 0.5 if e else None for e in ent]
err = [error(i) for i in x]

plt.style.use('dark_background')

fig = plt.figure()
ax = plt.subplot(111)

for i, lab, ls, c, in zip([ent, sc_ent, gini(x), err],
                          ['Entropy', 'Entropy (scaled)', 'Gini Impurity',
                                'Misclassification Error'],
                          ['-', '-', '--', '-.'],
                          ['white', 'purple', 'red', 'green', 'cyan'],):
    line = ax.plot(x, i, label=lab,
                   linestyle=ls, lw=2, color=c)
    
ax.legend(loc='upper center', bbox_to_anchor=(0.5, 1.15),
            ncol = 3, fancybox=True, shadow=False)
ax.axhline(y=0.5, linewidth=1, color='w', linestyle='--')
ax.axhline(y=1.0, linewidth=1, color='w', linestyle='--')

plt.ylim([0, 1.1])
plt.xlabel('p(i=1)')
plt.ylabel('Impurity Index')
plt.show()
```

With our theoretical basis in hand, let's train a decision tree. We'll have a maximum depth of 3 to avoid overfitting and we'll use *entropy* as our impurity criterion.

```python
from sklearn.tree import DecisionTreeClassifier

tree = DecisionTreeClassifier(criterion='entropy',
                              max_depth=3, random_state=0)

tree.fit(X_train,y_train)

X_combined = np.vstack((X_train, X_test))
y_combined = np.hstack((y_train, y_test))

plt.style.use('dark_background')

plot_decision_regions(X_combined, y_combined,
                      classifier=tree, test_idx=range(105,150))

plt.xlabel('petal length [cm]')
plt.ylabel('petal width [cm]')
plt.legend(loc='upper left')
plt.show()
```

The next topic covers a model that has gained popularity in recent years: the **Random Forest** model. The idea is to combine many **weak learners** to make a **strong learner** via a technique known as **ensemble learning**.

Random forests generally have good performance, scalability, and are relatively easy-to-use.

```python
%%time
from sklearn.ensemble import RandomForestClassifier

forest = RandomForestClassifier(criterion='entropy',
                                n_estimators=10,
                                random_state=1,
                                n_jobs=2) #only if you have a multi-core processor

forest.fit(X_train, y_train)

plt.style.use('dark_background')

plot_decision_regions(X_combined, y_combined,
                      classifier=forest, test_idx=range(105,150))

plt.xlabel('petal length [cm]')
plt.ylabel('petal width [cm]')
plt.legend(loc='upper left')
plt.title('Random Forest w/ 10 estimators')
plt.show()


                                
```

The final supervised learning algorithm we'll construct in this chapter is the **k-nearest neighbors classifier**. This algorithm is known as a *lazy learner* because it classifies by memorizing the dataset, not constructing a representative function.

As long as we have some notion of distance between points, we can create a KNN classifier that will predict the class label of a sample based on its closest k neighbors.

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=5, p=2,
                          metric='minkowski')

knn.fit(X_train_std, y_train)

plt.style.use('dark_background')
plot_decision_regions(X_combined_std, y_combined,
                      classifier=knn, test_idx=range(105,150))

plt.xlabel('petal length [standardized]')
plt.ylabel('petal width [standardized]')
plt.legend(loc='upper left')
plt.title('KNN Classifier based on 5 neighbors')
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=2, p=2,
                          metric='minkowski')

knn.fit(X_train_std, y_train)

plt.style.use('dark_background')
plot_decision_regions(X_combined_std, y_combined,
                      classifier=knn, test_idx=range(105,150))

plt.xlabel('petal length [standardized]')
plt.ylabel('petal width [standardized]')
plt.legend(loc='upper left')
plt.title('KNN Classifier based on 2 neighbors')
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=8, p=2,
                          metric='minkowski')

knn.fit(X_train_std, y_train)

plt.style.use('dark_background')
plot_decision_regions(X_combined_std, y_combined,
                      classifier=knn, test_idx=range(105,150))

plt.xlabel('petal length [standardized]')
plt.ylabel('petal width [standardized]')
plt.legend(loc='upper left')
plt.title('KNN Classifier based on 8 neighbors')
plt.show()
```

We see that our classifier is subject to our choice for the value of *k*.

KNN classifiers are notoriously subject to the **curse of dimensionality**. The more features we have, the higher dimensionality of our space and the more distances between points get expanded (see *Grus* page 160 for graph).
